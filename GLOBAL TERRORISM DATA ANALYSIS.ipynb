# **Correlation Analysis :**

```python
Corr_data = data[['Year','Month','Day','Country','State','Region','city','latitude','longitude','Attack_type','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(np.round(Corr_data.corr(),2),annot=True,cmap='PuRd')
```

This visualization is useful for quickly identifying patterns and relationships between variables in your dataset.

     *Positive correlations will appear in shades of red, with darker red indicating stronger positive correlations.
     
     *Negative correlations will appear in shades of purple, with darker purple indicating stronger negative correlations.
     
     *Values close to 0 will appear in lighter colors, indicating weak or no correlations.

# **Country Wise Terrorist Attacks:**

```python
fig,axes = plt.subplots(figsize=(16,11),nrows=1,ncols=2)
sns.barplot(x = data['Country'].value_counts()[:20].values, y = data['Country'].value_counts()[:20].index, 
            ax=axes[0],palette = 'coolwarm_r');
axes[0].set_title('Country Wise Terrorist Attacks')
sns.barplot(x=data['Region'].value_counts().values,y=data['Region'].value_counts().index,
            ax=axes[1])
axes[1].set_title('Region Wise Terrorist Attacks')
fig.tight_layout()
plt.show()
```

The resulting figure will have two bar plots side by side, one showing the top 20 countries with the most terrorist attacks and the other showing the number of terrorist attacks in each region. The colors of the bars in the first plot are based on the 'coolwarm_r' color palette, and the second plot uses the default Seaborn colors. This visualization allows you to quickly compare and analyze terrorist attack statistics by country and region.

# **Year Wise Terrorist Activities**

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',figsize=(10,5))
plt.title('Year wise Terrorist Activities by Region')
plt.ylabel('Number of Attacks')
plt.show()
```

The resulting plot will show a stacked area chart where each region is represented as a colored area, and the y-axis represents the number of terrorist attacks. The x-axis represents the years, and we can observe how the distribution of attacks across regions has changed over time. This visualization provides insights into the temporal patterns of terrorist activities across different regions.

```python
plt.bar(x='Year',height='Killed',data=New_data,color='orange')
plt.xticks(rotation=90)
plt.title('Year Wise Terrorist Activities')
```

```python
Year=New_data.Year.value_counts().to_dict()
rate=((Year[2017]-Year[1970])/Year[2017])*100
print('A total of', Year[1970],'attacks occured in 1970 and a total of',Year[2017],'attacks occured in the year 2017.')
print('The total number of attacks from 1970 to 2017 has increased by',np.round(rate,0),'percent.')
```

# **Attack methods used by Terrorists**

```python
# Create a Seaborn distribution plot
sns.displot(New_data['Attack_type'], color='red')

# Set x-axis label, rotate x-axis ticks, and set the title
plt.xlabel('Method')
plt.xticks(rotation=90)
plt.title('Methods of Attack')

# Set the figure size
plt.figure(figsize=(10, 5))

# Show the plot
plt.show()
```

```python
# Create a bar plot
plt.subplots(figsize=(8, 6))
sns.barplot(y=data['Group'].value_counts()[1:12].index, x=data['Group'].value_counts()[1:12].values, palette='muted')

# Set the title
plt.title('Most Active Terrorist Organizations')

# Show the plot
plt.show()
```

The resulting plot will show the top 11 most active terrorist organizations, with the number of incidents on the x-axis and the organization names on the y-axis. The bars will be colored using the 'deep' color palette.
